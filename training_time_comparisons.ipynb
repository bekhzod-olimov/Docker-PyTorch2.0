```python
import random, torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mnist_tr_times_2 = torch.load("times2.0/mnist_train_times")
mnist_val_times_2 = torch.load("times2.0/mnist_valid_times")

cifar10_tr_times_2 = torch.load("times2.0/cifar10_train_times")
cifar10_val_times_2 = torch.load("times2.0/cifar10_valid_times")

cifar100_tr_times_2 = torch.load("times2.0/cifar100_train_times")
cifar100_val_times_2 = torch.load("times2.0/cifar100_valid_times")

mnist_tr_times = torch.load("times1.13/mnist_train_times")
mnist_val_times = torch.load("times1.13/mnist_valid_times")

cifar10_tr_times = torch.load("times1.13/cifar10_train_times")
cifar10_val_times = torch.load("times1.13/cifar10_valid_times")

cifar100_tr_times = torch.load("times1.13/cifar100_train_times")
cifar100_val_times = torch.load("times1.13/cifar100_valid_times")
```

# MNIST Training

```python
index = np.arange(25)
bar_width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
summer = ax.bar(index, mnist_tr_times_2, bar_width,
                label="PyTorch 2.0")

winter = ax.bar(index+bar_width, mnist_tr_times,
                 bar_width, label="PyTorch 1.13")

ax.set_xlabel('Epochs')
ax.set_ylabel('Training Time (secs)')
ax.set_title('Training Times Comparison on the MNIST dataset')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels([a+1 for a in range(25)])
ax.legend()
plt.savefig('pngs/mnist_train.png', dpi=400)

plt.show()
```

```python
index = np.arange(25)
bar_width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))
summer = ax.bar(index, mnist_val_times_2, bar_width,
                label="PyTorch 2.0")

winter = ax.bar(index+bar_width, mnist_val_times,
                 bar_width, label="PyTorch 1.13")

ax.set_xlabel('Epochs')
ax.set_ylabel('Validation Time (secs)')
ax.set_title('Validation Times Comparison on the MNIST dataset')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels([a+1 for a in range(25)])
ax.legend()
plt.savefig('pngs/mnist_valid.png', dpi=400)

plt.show()
```
