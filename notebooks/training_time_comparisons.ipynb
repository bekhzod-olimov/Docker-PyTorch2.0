```python
import random, torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tr_times_2 = torch.load("stats/2.0_train_times")
val_times_2 = torch.load("stats/2.0_valid_times")

tr_times = torch.load("stats/1.0_train_times")
val_times = torch.load("stats/1.0_valid_times")

pl_tr_times_4 = torch.load("stats/pl_train_times_4_gpu")
pl_val_times_4 = torch.load("stats/pl_valid_times_4_gpu")

pl_tr_times_3 = torch.load("stats/pl_train_times_3_gpu")
pl_val_times_3 = torch.load("stats/pl_valid_times_3_gpu")

pl_tr_times_2 = torch.load("stats/pl_train_times_2_gpu")
pl_val_times_2 = torch.load("stats/pl_valid_times_2_gpu")

pl_tr_times_1 = torch.load("stats/pl_train_times_1_gpu")
pl_val_times_1 = torch.load("stats/pl_valid_times_1_gpu")

print(f"Average training time for PyTorch 1.12                 -> {np.mean(tr_times):.3f} secs.")
print(f"Average training time for PyTorch 2.01                 -> {np.mean(tr_times_2):.3f} secs.")
print(f"Average training time for PyTorch Lightning with 1 GPU -> {np.mean(pl_tr_times_1):.3f} secs.")
print(f"Average training time for PyTorch Lightning with 2 GPU -> {np.mean(pl_tr_times_2):.3f} secs.")
print(f"Average training time for PyTorch Lightning with 3 GPU -> {np.mean(pl_tr_times_3):.3f} secs.")
print(f"Average training time for PyTorch Lightning with 4 GPU -> {np.mean(pl_tr_times_4):.3f} secs.")
print("")
print(f"Average validation time for PyTorch 1.12                 -> {np.mean(val_times):.3f} secs.")
print(f"Average validation time for PyTorch 2.01                 -> {np.mean(val_times_2):.3f} secs.")
print(f"Average validation time for PyTorch Lightning with 1 GPU -> {np.mean(pl_val_times_1):.3f} secs.")
print(f"Average validation time for PyTorch Lightning with 2 GPU -> {np.mean(pl_val_times_2):.3f} secs.")
print(f"Average validation time for PyTorch Lightning with 3 GPU -> {np.mean(pl_val_times_3):.3f} secs.")
print(f"Average validation time for PyTorch Lightning with 4 GPU -> {np.mean(pl_val_times_4):.3f} secs.")
print("")
print(f"Total training time for PyTorch 1.12                 -> {np.sum(tr_times):.3f} secs.")
print(f"Total training time for PyTorch 2.01                 -> {np.sum(tr_times_2):.3f} secs.")
print(f"Total training time for PyTorch Lightning with 1 GPU -> {np.sum(pl_tr_times_1):.3f} secs.")
print(f"Total training time for PyTorch Lightning with 2 GPU -> {np.sum(pl_tr_times_2):.3f} secs.")
print(f"Total training time for PyTorch Lightning with 3 GPU -> {np.sum(pl_tr_times_3):.3f} secs.")
print(f"Total training time for PyTorch Lightning with 4 GPU -> {np.sum(pl_tr_times_4):.3f} secs.")
print("")
print(f"Total validation time for PyTorch 1.12                 -> {np.sum(val_times):.3f} secs.")
print(f"Total validation time for PyTorch 2.01                 -> {np.sum(val_times_2):.3f} secs.")
print(f"Total validation time for PyTorch Lightning with 1 GPU -> {np.sum(pl_val_times_1):.3f} secs.")
print(f"Total validation time for PyTorch Lightning with 2 GPU -> {np.sum(pl_val_times_2):.3f} secs.")
print(f"Total validation time for PyTorch Lightning with 3 GPU -> {np.sum(pl_val_times_3):.3f} secs.")
print(f"Total validation time for PyTorch Lightning with 4 GPU -> {np.sum(pl_val_times_4):.3f} secs.")
```

# Training Times

```python
epochs = 20
index = np.arange(epochs)
bar_width = 0.10

fig, ax = plt.subplots(figsize=(15, 7))

first = ax.bar(index, tr_times_2, bar_width, label="PyTorch 2.0")
second = ax.bar(index+bar_width, tr_times, bar_width, label="PyTorch 1.13")
third = ax.bar(index+2*bar_width, pl_tr_times_4, bar_width, label="PL Lightning 4 GPUs")
fourth = ax.bar(index+3*bar_width, pl_tr_times_3, bar_width, label="PL Lightning 3 GPUs")
fifth = ax.bar(index+4*bar_width, pl_tr_times_2, bar_width, label="PL Lightning 2 GPUs")
sixth = ax.bar(index+5*bar_width, pl_tr_times_1, bar_width, label="PL Lightning 1 GPU")

ax.set_xlabel('Epochs')
ax.set_ylabel('Training Time (secs)')
ax.set_title('Training Times Comparison')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels([a+1 for a in range(epochs)])
ax.legend()
plt.savefig('pngs/train_times.png', dpi=400)

plt.show()
```

# Validation Times

```python
epochs = 20
index = np.arange(epochs)
bar_width = 0.10

fig, ax = plt.subplots(figsize=(15, 7))

first = ax.bar(index, val_times_2, bar_width, label="PyTorch 2.0")
second = ax.bar(index+bar_width, val_times, bar_width, label="PyTorch 1.13")
third = ax.bar(index+2*bar_width, pl_val_times_4, bar_width, label="PL Lightning 4 GPUs")
fourth = ax.bar(index+3*bar_width, pl_val_times_3, bar_width, label="PL Lightning 3 GPUs")
fifth = ax.bar(index+4*bar_width, pl_val_times_2, bar_width, label="PL Lightning 2 GPUs")
sixth = ax.bar(index+5*bar_width, pl_val_times_1, bar_width, label="PL Lightning 1 GPU")

ax.set_xlabel('Epochs')
ax.set_ylabel('Validation Time (secs)')
ax.set_title('Validation Times Comparison')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels([a+1 for a in range(epochs)])
ax.legend()
plt.savefig('pngs/valid_times.png', dpi=400)

plt.show()
```
